```python
 #step 3: handling the missing value are data
 #scenario-1 missing value in numerical coloumn
 #calcuate the mean are average and replace the empty values
 #scenario-2 missing values in non numerical values
 #scenario-3 having missing value in huge dataset
 import numpy as np
 from sklearn.impute import SimpleImputer
 imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
 imputer.fit(x[:,1:3])
 x[:,1:3]=imputer.transform(x[:,1:3])


```

```python
#step-4 handling catogorial data
#which convert the text data into numeric data
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer#column transformer which is used to apply change of catogorical data to numerical data

#alexa-0,0,1

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')#encoder is method which convert text to binary
np.remainder='passthrough'
x=ct.fit_transform(x)
```
